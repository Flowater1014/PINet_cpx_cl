```python
import sys
sys.path.insert(0,'/root/autodl-tmp/PINet_cpx_cl')
import numpy as np 
import matplotlib.pyplot as plt
from src.config import PROJECT_ROOT
import os
```

```python
datasets_path = os.path.join(PROJECT_ROOT,'datasets','pinet_dataset_compared_7','train')
label_amp = np.load(os.path.join(datasets_path,'label_amp.npy'))
label_phs = np.load(os.path.join(datasets_path,'label_phs.npy'))
print(label_amp.dtype,label_amp.shape)
```

```python
def clip_percentile(arr,percent=0.5):
    low = np.percentile(arr,percent)
    high = np.percentile(arr,100-percent)
    return np.clip(arr,low,high)
n = 10
for i in range(n):
    amp = label_amp[i,:,:]
    phs = label_phs[i,:,:]
    fig,axes = plt.subplots(1,2,figsize=(10,4))
    axes[0].imshow(clip_percentile(amp),cmap='gray')
    axes[0].set_title('Amplitude')
    axes[0].axis('off')
    axes[1].imshow(clip_percentile(phs),cmap='gray')
    axes[1].set_title('Phase')
    axes[1].axis('off')
    plt.show()
    
```
